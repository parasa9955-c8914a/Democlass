```python
from transformers import AutoFeatureExtractor, AutoModelForAudioClassification

model_name = "Vladimirlv/ast-finetuned-audioset-10-10-0.4593-heart-sounds"

feature_extractor = AutoFeatureExtractor.from_pretrained(model_name)
model = AutoModelForAudioClassification.from_pretrained(
    model_name,
    num_labels=len(unique_labels),
    ignore_mismatched_sizes=True
)
```

```python
import torch

device = "cuda" if torch.cuda.is_available() else "cpu"
print(f"Using device: {device}")

def preprocess_function(examples):
    # Extract the 'array' from the Audio feature
    audio_arrays = [x["array"] for x in examples["filename"]]
    inputs = feature_extractor(
        audio_arrays,
        sampling_rate=feature_extractor.sampling_rate,
        max_length=16000,
        truncation=True,
        padding="max_length",
        return_tensors="pt"
    )
    return inputs

# Map the dataset. Note: Audio decoding happens on CPU.
# We'll remove the original 'filename' column to save memory after extraction.
# token removed
# token removed

# Move model to GPU
model.to(device)
```

```python
from transformers import TrainingArguments, Trainer
import numpy as np
from evaluate import load

metric = load("accuracy")

def compute_metrics(eval_pred):
    logits, labels = eval_pred
    predictions = np.argmax(logits, axis=-1)
    return metric.compute(predictions=predictions, references=labels)

training_args = TrainingArguments(
    output_dir="./results",
    eval_strategy="epoch",
    save_strategy="epoch",
    learning_rate=3e-5,
    per_device_train_batch_size=10,
    per_device_eval_batch_size=10,
    num_train_epochs=5,
    warmup_ratio=0.1,
    logging_dir="./logs",
    logging_steps=10,
    load_best_model_at_end=True,
    metric_for_best_model="accuracy",
)

trainer = Trainer(
    model=model,
    args=training_args,
# token removed
# token removed
    compute_metrics=compute_metrics,
)

trainer.train()
```

```python
# data size

import os
import pandas as pd


folder_path = "/content/dataset/the-circor-digiscope-phonocardiogram-dataset-1.0.3/training_data"

wav_files = []
data = []

file_count = 0
total_size = 0

for root, dirs, files in os.walk(folder_path):

    for file in files:

        if file.lower().endswith(".wav"):

            full_path = os.path.join(root, file)

            wav_files.append(full_path)

            file_count += 1

            total_size += os.path.getsize(full_path)

            label = os.path.basename(root)

            data.append({

                "fname": full_path,
                "label": label

            })

df = pd.DataFrame(data)

print(df.head())


total_size_mb = total_size / (1024 * 1024)

print("=== DATASET SIZE ===")

print("Total wav files:", len(wav_files))

print("Total size:", round(total_size_mb, 2), "MB")
```

```python
import pandas as pd
import os

data = []

for wav_path in wav_files:

    file_name = os.path.basename(wav_path)

    data.append({

        "filename": file_name,
        "path": wav_path

    })

df = pd.DataFrame(data)
print(df.columns)
```

```python
# convert name

def convert_label(file_name):

    upper_name = file_name.upper()

    if (
        "AV" in upper_name
        or "MV" in upper_name
        or "PV" in upper_name
        or "TV" in upper_name
    ):

        return "murmur"

    else:

        return "normal"

df["label"] = df["filename"].apply(convert_label)
print(df.head())
```
